```python
# Importing pandas library
import pandas as pd
```

```python
#Files to load
school_data_to_load="Resources/schools_complete.csv"
student_data_to_load="Resources/students_complete.csv"
```

```python
#Creating schools data dataframe
school_data_df=pd.read_csv(school_data_to_load)
school_data_df
```

```python
#Creating students data dataframe
student_data_df=pd.read_csv(student_data_to_load)
student_data_df
```

```python
#Determining missing values in students data using count
student_data_df.count()
```

```python
# Determining datatype for a column
school_data_df["school_name"].dtype
```

```python
# Determining datatype for student data
student_data_df.dtypes
```

```python
# Data cleanup to remove prefixes and suffixes
# List of possible prefixes and suffixes
prefix_suffix=["Dr. ","Mr. ","Ms. ","Mrs. ","Miss "," DDS"," DVM"," MD"," PhD"]
# Replacing the prefixes and suffixes with blank by iterating through the list
for item in prefix_suffix:
    student_data_df["student_name"]=student_data_df["student_name"].str.replace(item,"")
student_data_df
```

```python
#Merging dataframes
school_data_complete_df=pd.merge(student_data_df,school_data_df,on=["school_name",'school_name'])
school_data_complete_df.head()
```

```python
#Total number of students
student_count=school_data_complete_df["Student ID"].count()
student_count
```

```python
#Total number of schools using school data df
school_count=school_data_df["School ID"].count()
school_count
```

```python
#Total number of schools using the merged df
school_count_2=len(school_data_complete_df["School ID"].unique())
school_count_2
```

```python
#Total budget for district
total_budget=school_data_df["budget"].sum()
total_budget
```

```python
#Average math scores
average_math_score=school_data_complete_df["math_score"].mean()
average_math_score
```

```python
#Average reading scores
average_reading_score=school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
#Calculate students who passed math and store in new df
passing_math=school_data_complete_df[school_data_complete_df["math_score"]>=70]
passing_math.head

```

```python
#Calculate students who passed reading and store in new df
passing_reading=school_data_complete_df[school_data_complete_df["reading_score"]>=70]
passing_reading.head()
```

```python
# Calculate number of students who passed math
passing_math_count=passing_math["Student ID"].count()
passing_math_count
```

```python
# Calculate number of students who passed reading
passing_reading_count=passing_reading["Student ID"].count()
passing_reading_count
```

```python
passing_grade_math=school_data_complete_df["math_score"]>=70
passing_grade_math.sum()
```

```python
# Calculate percentage that passed math
passing_math_percentage=float(passing_math_count/student_count) *100
passing_math_percentage
```

```python
# Calculate percentage that passed reading
passing_reading_percentage=float(passing_reading_count/student_count) *100
passing_reading_percentage
```

```python
# Calculate percentage of students who passed both math and reading
passing_math_reading=school_data_complete_df[(school_data_complete_df["reading_score"]>=70) & (school_data_complete_df["math_score"]>=70)]
passing_math_reading_count=passing_math_reading["Student ID"].count()
overall_passing_percentage=passing_math_reading_count/student_count *100
overall_passing_percentage
```

```python
# Adding results to a new DF
district_summary_df=pd.DataFrame([{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": total_budget,
          "Average Math Score": average_math_score,
          "Average Reading Score": average_reading_score,
          "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])
district_summary_df
```

```python
district_summary_df["Total Students"]=district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"]=district_summary_df["Total Budget"].map("${:,}".format)
district_summary_df["Average Math Score"]=district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["Average Reading Score"]=district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["% Passing Math"]=district_summary_df["% Passing Math"].map("{:.0f}".format)
district_summary_df["% Passing Reading"]=district_summary_df["% Passing Reading"].map("{:.0f}".format)
district_summary_df["% Overall Passing"]=district_summary_df["% Overall Passing"].map("{:.0f}".format)
district_summary_df
```
